This notebook plots out what power-transformed features look like compared to original values

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import power_transform
```

```python
DATA_DIR = Path.cwd().parent / "train-test"
```

```python
df_train = pd.read_parquet(DATA_DIR/"train.parquet")
df_test = pd.read_parquet(DATA_DIR/"test.parquet")

print(f"Shape of the training data : {df_train.shape}")
print(f"Shape of the test data : {df_test.shape}")

X_train = df_train.iloc[:,1:-1]
y_train = df_train.iloc[:,-1]

X_test = df_test.iloc[:,1:-1]
y_test = df_test.iloc[:,-1]

print(f"Training proportion: {len(y_train) / (len(y_train) + len(y_test))}")
```

```python
def plot_tf(feature):
    plt.hist(feature); plt.title("Raw"); plt.show()
    plt.hist(power_transform(np.reshape(feature.values, (-1,1)))); plt.title("Power"); plt.show()
```

```python
plot_tf(X_train.r_length)
```

```python
plot_tf(X_train.u_friends_count)
```

```python
plot_tf(X_train.u_review_count)
```

```python
plot_tf(X_train.u_month_age)
```

```python
plot_tf(X_train.b_review_count)
```

```python
plot_tf(X_train.r_rea)
```
